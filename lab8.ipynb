```python
import os
# 1 & 2: Create folders
if not os.path.exists("lab_data"):
    os.makedirs("lab_data")

for sub in ["input", "output", "logs"]:
    path = os.path.join("lab_data", sub)
    if not os.path.exists(path):
        os.makedirs(path)

# 3: List items and check type
items = os.listdir("lab_data")
for item in items:
    path = os.path.join("lab_data", item)
    if os.path.isdir(path):
        print(item, "is a directory")
    else:
        print(item, "is a file")

# 4: Rename logs to history
old_p = os.path.join("lab_data", "logs")
new_p = os.path.join("lab_data", "history")
if os.path.exists(old_p):
    os.rename(old_p, new_p)
```

```python
import shutil
# 5: Create and write 5 lines
file_path = os.path.join("lab_data", "input", "notes.txt")
with open(file_path, "w") as f:
    f.write("Line 1\nLine 2\nLine 3\nLine 4\nLine 5\n")

# 6: Read and count lines
with open(file_path, "r") as f:
    print("Total lines:", len(f.readlines()))

# 7: Copy file
dest = os.path.join("lab_data", "output", "summary.txt")
shutil.copy(file_path, dest)
```
